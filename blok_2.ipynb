```python
def simpson(a,b, y):
    h = (b-a)/2
    return (h/3)*(y[0]+4*y[1]+y[2])

def trapezoid(a,b, y):
    h = b - a
    return (h/2)*(y[0]+y[1])

print(simpson(1,2,[10,7,5]))
print(trapezoid(1,2,[10,2]))
```

```python
def comp_trap(a,b,n,f):
    h = (b-a)/n
    y = [f(a+i*h) for i in range(n)]
    edge_vals = 0.5*(y[0]+y[-1])
    return round(h*(edge_vals+sum(y[1:-1])),2)
f = lambda x: math.sin(x)
comp_trap(0,math.pi,20,f)
```
